```julia
using Combinatorics, Plots
Plots.GRBackend()
```

```julia
function calc(n,p,k)
    probs=[]
    exp_cases=0
    max_prob=0
    total_prob_scaling=0
    prob_greater=0;
    for i in k:n-p+k
        prob=binomial(big(i),big(k))*binomial(big(n-i),big(p-k))/binomial(big(n),big(p))
        append!(probs,prob)
        exp_cases+=prob*i
        total_prob_scaling+=prob
    end
    exp_cases/=total_prob_scaling
    prob_greater=sum(probs[convert(Int64,ceil(exp_cases)+1-k):n-p])
    #est_cases=k*n/p
    probs./total_prob_scaling
    return(exp_cases,prob_greater/total_prob_scaling,probs./total_prob_scaling);
end
```

```julia
plot(1:190,(1:190)./1000000000000,label=false)
println("N=200, tested=10")
for i in 1:4
    println("k = ",i,"\nExp cases : ",Float32(calc(200,10,i)[1]),"\nP (cases > exp_cases+1 | k = ",i,") = ",Float32(calc(200,10,i)[2]),"\n","Estimated cases : ",i*20,"\n")
    plot!(calc(200,10,i)[3],label="k=$i")
end
plot!(1:190,(1:190)./1000000000000,label=false)
```
