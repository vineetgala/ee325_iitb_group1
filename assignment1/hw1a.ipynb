```julia
using StatsBase
using Statistics
using Plots
```

```julia
function scene1(k,data)
    sample_avgs=[]
    arbit=0
    sample_stddevs=[]
    for i in 1:50
        arbit=sample(1:size(data)[1]-k)
        sample_data=data[arbit:arbit+k-1]
        sample_avg=mean(sample_data)
        sample_stddev=std(sample_data)
        push!(sample_stddevs,sample_stddev)
        push!(sample_avgs,sample_avg)
    end
    #println("Standard deviation of means from scenario-1 : ",std(sample_avgs))
    #println("Mean of standard deviations from scenario-1 : ",mean(sample_stddevs))
    #println("Mean from scenario-1 : ",mean(sample_avgs))
    scatter([1:50],sample_avgs,label="scenario-1")
end
```

```julia
function scene2(k,data)
    sample_avgs=[]
    sample_stddevs=[]
    for i in 1:50
        sample_data=data[convert(Int64,(i-1)*10000/50+1):convert(Int64,(i-1)*10000/50+k)]
        sample_avg=mean(sample_data)
        sample_stddev=std(sample_data)
        push!(sample_avgs,sample_avg)
        push!(sample_stddevs,sample_stddev)
    end
    #println("Standard deviation of means from scenario-2 : ",std(sample_avgs))
    #println("Mean of standard deviations from scenario-2 : ",mean(sample_stddevs))
    #println("Mean from scenario-2 : ",mean(sample_avgs))
    scatter([1:50],sample_avgs,label="scenario-2")
end
```

```julia
function scene3(k,data)
    sample_avgs=[]
    sample_stddevs=[]
    for i in 1:50
        random=sample(1:10000,k)
        sample_data=[]
        for j in 1:k
            push!(sample_data,data[random[j]])
        end
        sample_avg=mean(sample_data)
        sample_stddev=std(sample_data)
        push!(sample_avgs,sample_avg)
        push!(sample_stddevs,sample_stddev)
    end
    #println("Standard deviation of means from scenario-3 : ",std(sample_avgs))
    #println("Mean of standard deviations from scenario-3 : ",mean(sample_stddevs))
    #println("Mean from scenario-3 : ",mean(sample_avgs))
    scatter([1:50],sample_avgs,label="scenario-3")
end
```

```julia
open("Assignment 1/hw1a.txt") do f
    data=[]
    while !eof(f)
        push!(data,parse(Float64,readline(f)))
    end
    avg=mean(data)
    stddev=std(data)
    #println("Standard Deviation : ",stddev)
    #println("Mean : ",avg)
    gr()
    scatter()
    k=10
    p=[]
    push!(p,scene1(k,data))
    push!(p,scene2(k,data))
    push!(p,scene3(k,data))
    scatter(p[1],p[2],p[3],layout=(3,1),title="K=$k")
end
```
