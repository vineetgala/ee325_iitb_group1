```julia
using StatsBase
using Statistics
using Plots
```

```julia
data=[[], [], [], []]
for i in 1:4
    open("Assignment 1/hw1b"*string(i)*".txt") do f
        data1=[]
        while !eof(f)
            push!(data[i],parse(Bool,readline(f)))
        end
    end
end
```

```julia
data=convert(Array{Array{UInt8}}, data)
```

```julia
mean_data=[[],[],[],[]]
std_data=[[],[],[],[]]
exp_ub_data1=[[],[],[],[]]
exp_lb_data1=[[],[],[],[]]
exp_ub_data2=[[],[],[],[]]
exp_lb_data2=[[],[],[],[]]
factor1=1
factor2=1.9
for i in 1:length(data)
    for j in 1:length(data[i])
        push!(std_data[i],std(data[i][1:j])/(j^0.5))
        push!(mean_data[i],mean(data[i][1:j]))
        push!(exp_ub_data1[i],0.5+factor1/(2*sqrt(j)))
        push!(exp_lb_data1[i],0.5-factor1/(2*sqrt(j)))
        push!(exp_ub_data2[i],0.5+factor2/(2*sqrt(j)))
        push!(exp_lb_data2[i],0.5-factor2/(2*sqrt(j)))
    end
end
```

```julia
i=1
gr()
plot([1:length(mean_data[i])], mean_data[i])
plot!([1:length(mean_data[i])], exp_ub_data1[i])
plot!([1:length(mean_data[i])], exp_lb_data1[i])
plot!([1:length(mean_data[i])], exp_ub_data2[i])
plot!([1:length(mean_data[i])], exp_lb_data2[i])
```
