```julia
# importing required libraries
using StatsBase
using Statistics
using Plots
```

```julia
# reading all 4 files
data=[[], [], [], []]
for i in 1:4
    open("Assignment 1/hw1b"*string(i)*".txt") do f
        data1=[]
        while !eof(f)
            push!(data[i],parse(Bool,readline(f)))
        end
    end
end
```

```julia
# converting to integer format
data=convert(Array{Array{UInt8}}, data)
```

```julia
# calculating mean and std of mean at each time step
mean_data=[[],[],[],[]]
std_data=[[],[],[],[]]
for i in 1:length(data)
    for j in 2:length(data[i]) # since mean and std make no sense for 1 data point
        push!(std_data[i],std(data[i][1:j])/(j^0.5))
        push!(mean_data[i],mean(data[i][1:j]))
    end
end
```

```julia
# checking whether 0.5 is between the two bounds for the actual mean based on k-factor
```

```julia
i=1
k=2
gr()
plot([1:length(mean_data[i])], ones(length(mean_data[i])).*0.5)
plot!([1:length(mean_data[i])], mean_data[i]+k*std_data[i])
plot!([1:length(mean_data[i])], mean_data[i]-k*std_data[i])
```
