```julia
using Plots
using Statistics
```

```julia
url_3 = raw"C:\Users\Rohit\Downloads\hw1c1.txt"
```

```julia
f_3 = open(url_3, "r")
```

```julia
s = read(f_3, String);
```

```julia
close(f_3)
```

```julia
hw = split(s, "\n")
pop!(hw)
x,y = [], []
for i in hw
    j, k = split(i, "\t")
    push!(x, parse(Float64, j))
    push!(y, parse(Float64, k))
end
```

```julia
x_bar, y_bar = sum(x)/length(x), sum(y)/length(y)
```

```julia
a = sum(@.((x-x_bar)*(y-y_bar)))/sum(@.((x-x_bar)^2))
```

```julia
b = y_bar - a * x_bar
```

```julia
plot(x, y, seriestype = :scatter, label = "Actual")
y_pred = @.(a*x + b)
plot!(x, y_pred, label = "Predicted", title = "Scatter Plot", xlabel = "Height[cms]", ylabel = "Weight[Kgs]")
```
