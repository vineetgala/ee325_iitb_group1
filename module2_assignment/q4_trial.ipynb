```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 50
```

```python
def matrix_elements(i, j):
    if abs(i-j)>1:
        return 0
    if i==j:
        return (2 * i * (n-i)) / (n * n)
    if i>j:
        if j==n:
            return 0
        return ((j+1)/n)**2
    if j==0:
        return 0
    return ((n+1-j)/n)**2
```

```python
def iterator():
    for i in range(n+1):
        for j in range(n+1):
            yield matrix_elements(i, j)
```

```python
iterable = [x for x in iterator()]
```

```python
mat=np.fromiter(iterable, float)
```

```python
mat=mat.reshape((n+1, n+1))
```

```python
x_0=np.zeros((1, n+1))
x_0[0, n]=1
x_0
```

```python
def x_m(m):
    mat_new = np.linalg.matrix_power(mat, m)
    return x_0 @ mat_new
```

```python
for m in [1, 5, 10, 20, 50, 100, 500]:
    plt.bar(range(n+1), x_m(m).reshape(-1))
    plt.title("m = "+str(m))
    plt.show()
```
